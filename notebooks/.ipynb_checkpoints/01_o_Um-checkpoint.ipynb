## 1. A gênese da contagem

Tudo o que chamamos de número nasce de um ato: **repetir o 1**. Dele surge a sequência natural:

$$
\mathbb{N} = \{1, 2, 3, 4, \dots\}
$$

A **Mônada** — o **1** — não é apenas o primeiro elemento dessa sequência, mas o princípio que a permite existir. Contar é, essencialmente, repetir o 1 em intervalos iguais.

A partir desse ato simples, o mundo inteiro das estruturas discretas emerge. E entre essas estruturas, há uma que resiste à decomposição: os **números primos**. Eles são as *partes indecomponíveis do Um* quando a unidade é repetida e combinada.

---

## 2. A Dobra

Considere a reta numérica de $1$ a $x$ e **dobre-a ao meio**, no ponto $x/2$. Essa operação revela uma simetria simples e fértil: os primos se separam naturalmente em dois papéis complementares.

- **Estruturadores** — os primos em $[1, x/2]$: são os “blocos” básicos. Por estarem na primeira metade, seus múltiplos $2p,3p, \cdots$ ainda caem dentro do intervalo observado $[1, x]$ e, portanto, **estruturam** a população dos números compostos.

- **Estabilizadores** — os primos em $(x/2, x]$: mais recentes. Como $p \gt x/2$, já não conseguem criar novos compostos em $[1, x]$ (pois $2p \gt x$). Seu papel é **estabilizar** a sequência observada, ocupando as lacunas deixadas pelos compostos.

Essa dualidade nos motiva a definir um *sismógrafo aritmético* — uma metáfora observacional para uma função que mede, a cada instante $x$, o desbalanceamento entre os dois grupos de primos. A medição é uma comparação direta entre a quantidade de **primos estabilizadores** e a de **primos estruturadores**.

Para expressar isso matematicamente, primeiro definimos a função $\pi(x)$ (lê-se "pi de $x$"), que simplesmente conta o número total de primos menores ou iguais a $x$.

Com isso, o nossa função-sismógrafo, que chamamos de $\Delta_\pi(x)$, é formalizada como a diferença:

$$
\Delta_\pi(x)
=
\underbrace{\big(\pi(x) - \pi(\lfloor x/2 \rfloor)\big)}_{\text{Estabilizadores}}
-
\underbrace{\pi(\lfloor x/2 \rfloor)}_{\text{Estruturadores}}
$$

Simplificando a expressão, chegamos à sua forma final e compacta:

$$
\Delta_\pi(x) = \pi(x) - 2 \cdot \pi(\left\lfloor \frac{x}{2} \right\rfloor)
$$

> **Intuição-chave**: tudo nasce do gesto de dobrar a reta — separar quem ainda “constrói” compostos de quem apenas “preenche” vazios.

Neste estágio, essa função aritmética é apenas definida. Nenhuma análise estatística ou espectral é realizada aqui.

## 3. Visualização: Os Dois Papéis em Ação

Vamos visualizar os primos até $N$, colorindo:

- em **azul**, os primos de $[1, N/2]$ (*estruturadores*);

- em **laranja**, os primos de $(N/2, N]$ (*estabilizadores*).

Na legenda, exibiremos $\pi(N/2)$ e $\pi(N) - \pi(N/2)$, respectivamente, para quantificar cada papel.


```python
import numpy as np
import matplotlib.pyplot as plt

# Cores
CIP_BLUE   = "#0055A4"  # estruturadores em [1, x/2]
CIP_ORANGE = "#FF7F11"  # estabilizadores em (x/2, x]

def primos_ate(n: int) -> np.ndarray:
    if n < 2: return np.array([], dtype=int)
    sieve = np.ones(n+1, dtype=bool); sieve[:2] = False
    for i in range(2, int(np.sqrt(n))+1):
        if sieve[i]: sieve[i*i:n+1:i] = False
    return np.flatnonzero(sieve)

def plot_primos_dobra_1linha(N: int = 100, s=28):
    p = primos_ate(N)
    mid = N//2
    p_left  = p[p <=  mid]          # estruturadores
    p_right = p[p >   mid]          # estabilizadores

    fig, ax = plt.subplots(figsize=(10, 2.2))

    # única “reta”: y=0
    y0 = np.zeros_like(p)
    ax.scatter(p_left,  y0[:p_left.size],  s=s, color=CIP_BLUE,   label=f'[1, x/2] — "estruturadores": {p_left.size}',  zorder=3)
    ax.scatter(p_right, y0[:p_right.size], s=s, color=CIP_ORANGE, label=f'(x/2, x] — "estabilizadores": {p_right.size}', zorder=3)

    # guias visuais
    ax.axhline(0, color="0.85", lw=1)
    ax.axvline(mid, color="0.85", lw=1)

    # estética
    ax.set_yticks([])                 # só a reta
    ax.set_xlim(-1, N+1)
    ax.set_xlabel("n")
    ax.set_title(f"Dobra dos primos (N = {N})")
    ax.legend(loc="upper center", ncol=2, frameon=False, bbox_to_anchor=(0.5, -0.25))
    plt.tight_layout()
    plt.show()

# experimente:
plot_primos_dobra_1linha(100)
```

### **Experimento**  
Altere o valor de $N$ para $200$, $400$, $800$. Observe como os pontos se distribuem ao longo das duas metades. A estrutura permanece — apenas se replica em escala.

O leitor atento perceberá algo essencial: a cada dobra, parte da informação de $[1, x / 2]$ se reflete em $(x / 2, x]$ através da estruturação dos números compostos. Nada é acrescentado de fora. O sistema se auto-estrutura.

**Nota:** o valor $1$ não desempenha papel operacional nas relações observadas. Ele sustenta a contagem, mas não participa das conexões estruturais reveladas pela dobra.

Neste modelo, o Um não atua como elemento relacional. Ele define o passo da contagem, não as relações entre posições.

O Dois é o primeiro espelho da Unidade e inaugura a relação observável. A partir dele, a dobra torna-se possível.

Os vazios da dobra são medidos em unidades de $1$, mas apenas números maiores que $1$ ocupam posições estruturais.
